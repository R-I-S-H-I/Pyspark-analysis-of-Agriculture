```python
from pyspark.sql import DataFrame
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
from pyspark.sql.types import StructType, StringType, IntegerType, StructField, DoubleType


spark = SparkSession.builder.appName("temp_file_pyspark").getOrCreate()
```

```python
df_csv_file_agriculture = spark.read.option("InferSchema","True").option("Header","True").csv("/content/crop_yield.csv")
```

```python
df_csv_file_agriculture.toPandas().head(5)
```

```python
df_csv_file_agriculture.printSchema()
```

```python
myView = StructType([StructField('Region', StringType(), True), StructField('Soil_Type', StringType(), True), StructField('Crop', StringType(), True), StructField('Rainfall_mm', DoubleType(), True)])
```

```python
data_of_csv_file = spark.read.option("InferSchema","True").option("Header","True").schema(myView).csv("/content/crop_yield.csv")
```

```python
output = data_of_csv_file.select("*").withColumn("insrt_time", F.current_timestamp()).where((data_of_csv_file.Region == "North") & (data_of_csv_file.Rainfall_mm >900))
```

```python
pd_df = output.toPandas()
```
